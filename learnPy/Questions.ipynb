```python
#27
class Solution(object):
    def removeElement(self, nums, val):
        """
        :type nums: List[int]
        :type val: int
        :rtype: int
        """
        if len(nums) == 0:
            return 0
        end = len(nums) - 1
        start = 0
        res = 0
        while end > start:
            if nums[start] == val:
                res += 1
                while end > start and nums[end] == val:
                    res += 1
                    print(res)
                    end -= 1
                    if end == start:
                        return len(nums) - res
                nums[start] = nums[end]
                start += 1
                end -= 1
            else:
                start += 1
        if nums[start] == val:
            res += 1
        return len(nums) - res
```

```python
# 150
'1'.isnumeric()
# time complexity of x in [1,2,3]

class Solution(object):
    def evalRPN(self, tokens):
        """
        :type tokens: List[str]
        :rtype: int
        """
        s = []
        op = ["+", "-", "*", "/"]
        for c in tokens:
            if c not in op:
                s.append(int(c))
            else:
                o2 = s.pop()
                o1 = s.pop()
                if c == '+':
                    s.append(o1+o2)
                elif c == '-':
                    s.append(o1-o2)
                elif c == '*':
                    s.append(o1*o2)
                else:
                    s.append(int(o1/o2))
                print(s)
        
        return s.pop()
```

```python
# 102
class Solution(object):
    def levelOrder(self, root):
        """
        :type root: TreeNode
        :rtype: List[List[int]]
        """
        if not root:
            return []
        hq = [root]
        heapq.heapify(hq)
        res = []
        while hq:
            temp_res = []
            for i in range(len(hq)):
                r = heapq.heappop(hq)
                print(r)
                temp_res.append(r)
                heapq.heappush(hq, r.left)
                heapq.heappush(hq, r.right)
            res.append(temp_res)
        return res
```

```python
class Solution(object):
    def partition(self, s):
        """
        :type s: str
        :rtype: List[List[str]]
        """
        res = []
        path = [0]
        element_res = []
        def cp(st):
            si = 0
            e = len(st)-1
            while si<e:
                if st[si] != st[e]:
                    return False
                si += 1
                e -= 1
            return True
        def bt(start):
            if start > len(s):
                res.append(element_res[:])
                return
            
            for i in range(start, len(s)+1):
                if not cp(s[path[-1]:i]):
                    continue
                element_res.append(s[path[-1]:i])
                print(i,element_res)
                path.append(i)
                bt(i+1)
                path.pop()
                element_res.pop()
        bt(1)
        return res
```

```python
# 122
class Solution(object):
    def maxProfit(self, prices):
        """
        :type prices: List[int]
        :rtype: int
        """
        pre_price = int(prices[0])
        profit = 0
        hold = True
        prices.append(-1)
        for i in range(len(prices)-1):
            if hold:
                if prices[i] > prices[i+1]:
                    print('hold', profit, prices[i], pre_price)
                    profit = profit + prices[i] - pre_price
                    hold = False
            else:
                if prices[i] < prices[i+1]:
                    print('no hold', prices[i], pre_price)
                    pre_price = int(prices[i])
                    print('no hold', prices[i], pre_price)
                    hold = True
        return profit
```

```python
class Solution(object):
    def findLength(self, nums1, nums2):
        """
        :type nums1: List[int]
        :type nums2: List[int]
        :rtype: int
        """
        dp = [[0 for i in range(len(nums2))] for j in range(len(nums1))]
        
        for i in range(len(nums1)):
            if nums1[i] == nums2[0]:
                for j in range(i, len(nums1)):
                    dp[j][0] = 1
                break
        
        for i in range(len(nums2)):
            if nums2[i] == nums1[0]:
                for j in range(i, len(nums2)):
                    dp[0][j] = 1
                break
        
        for col in range(1, len(nums2)):
            for row in range(1, len(nums1)):
                if nums1[row] == nums2[col]:
                    dp[row][col] = dp[row-1][col-1] + 1
                    for i in range(1, min(row, col) - dp[row-1][col-1] + 1):
                        if nums1[row-i] != nums2[col-i]:
                            dp[row][col] = max(dp[row-1][col-1],1,dp[row][col-1], dp[row-1][col])
                            break
                else:
                    dp[row][col] =  max(dp[row-1][col-1], dp[row][col-1], dp[row-1][col])
        for k in dp:
            print(k)
        return dp[-1][-1]
```

```python
class Solution(object):
    def buildTree(self, inorder, postorder):
        """
        :type inorder: List[int]
        :type postorder: List[int]
        :rtype: TreeNode
        """
        
        def recur(inorder, postorder, si, ei, sp, ep):
            if ep == sp:
                return None
            
            ep -= 1
            root_val = postorder[ep]
            
            if sp + 1 == ep:
                return TreeNode(root_val)
            root_index = 0

            for i in range(ei - si + 1):
                if root_val == inorder[i]:
                    root_index = i
                    break
            print(inorder[si: si+1])
            print(postorder[sp: sp+i])
            print(inorder[si + i + 1: ei])
            left_node = recur(inorder, postorder, si, si + i, sp, sp + i )
            right_node = recur(inorder, postorder, si + i + 1, ei, sp + i, ep)
            
            return TreeNode(root_val, left_node, right_node)
        
        return recur(inorder, postorder, 0, len(inorder), 0, len(postorder))
```
